1. Change the data and see its impact. You can change it by adding some gaussian noise using the np.random.randn function. See how this affects the scatter plot, the covariance matrix, the scree plot and the principal components and their Eigen values.

```python
import numpy as np
weights = np.array([55,74,58,52,58,67,80,71,62,69,72,67,78])
heights = np.array([56, 73, 60, 54, 56, 62, 77, 73, 72, 69, 74, 67, 84,])
```

Adding some noise

```python
n_weights = np.random.normal(4*weights+10,25)
n_heights = heights
```

Plotting

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(n_heights,n_weights)
plt.title('New Plot')
plt.xlabel('Heights')
plt.ylabel('Weights')
plt.grid()

plt.figure()
plt.scatter(heights,weights)
plt.title('Original Plot')
plt.xlabel('Heights')
plt.ylabel('Weights')
plt.grid()
plt.show()
```

```python
X = np.vstack((n_heights,n_weights)).T
Xm = np.mean(X,axis=0)
Xs = np.std(X,axis=0)

Xn = (X-Xm)/(Xs)
```

# Covariance matrix

```python
C = np.cov(Xn.T)
print("Covariance matrix is\n",C)
```

Original Covariance matrix was:  [[1.08333333 0.98631656]
 [0.98631656 1.08333333]]

# Eigen Values and Eigen Vectors

```python
ev,pc = np.linalg.eig(C)
idx = np.argsort(-ev)
ev = ev[idx]
pc = pc[:,idx]
print("The eigen values are",ev)
print("The eigen vectors (principle components) are \n",pc)
```

Orgininal Eigen Values and Eigen Vectors were: 
The eigen values: [2.0696499  0.09701677]

The eigen vectors (principle components): 
 [[-0.70710678 -0.70710678]
 [-0.70710678  0.70710678]]

## Scree Plot


```python
plt.plot(np.cumsum(ev)/np.sum(ev),'o-')
plt.xlabel("number of components")
plt.ylabel("Fraction of variance")
plt.grid()
plt.title("Scree plot")
```

## 2. Can you overlay the principal components of your data as vectors over the scatter plot to see whether the direction of maxium variance being found is actually correct.

```python
dr = 1
wx = pc[:,0:dr]
z = np.dot(Xn,wx)

DR  = 2
WX = pc[:,1:DR]
Z = np.dot(Xn,WX)

print("The principle component matrix multiple by its transpose:\n",np.dot(pc,pc.T))
iwx = pc.T[:,0:dr] 
Xr = np.dot(z,iwx.T)*Xs+Xm
plt.scatter(Xr[:,0],Xr[:,1],marker='^')

IWX = pc.T[:,1:DR] 
XR = np.dot(Z,IWX.T)*Xs+Xm
plt.scatter(XR[:,0],XR[:,1], marker='*')

plt.scatter(X[:,0],X[:,1])
plt.grid()
plt.show()
```

So here Green Circles are showing original points(n_heights,n_weights).
Blue hat points are showing points where variance of data will be maximum which is also maximum principal component.
Yellow stars are showing points where variance of data will be minimum which is also minimum principal component.

So it is clear from figure that direction of maximum variance is correct.
